## 1. 데이터 시각화 이해

### 1.1 데이터 시각화 장점
* 기술적인 데이터 분석 결과 기반의 가치 창출이 데이터 분석의 핵심
* 데이터 시각화를 통해 데이터의 인사이트를 발견하고, 스토리텔링이 용이해짐

### [많은 양의 데이터를 시각적으로 요약]
* **도형, 선, 색 등의 시각적 요소들을 활용**하여 많은 양의 데이터를 요약 표현
* 의료, 금융, 교통 등 데이터가 생성되는 대부분의 분야에서 **시각화 없이는 현상을 파악하고 예측하기 어렵기 때문에** 데이터 시각화가 강조됨
* 어떤 데이터를 어떻게 표현할 것인지에 따라 **다양한 시각화 유형이 존재**



### [시각을 통한 데이터 인사이트 도출]
* 시각화 차트에서 시각화 요소인 도형의 형태, 크기, 위치, 색 정보 등으로 **시각적 패턴 도출**
* 데이터 인사이트의 정확한 전달을 위해 **시각화 형태를 잘 선택하고 데이터 시각화의 목적을 명확히 해야 함**
  - **데이터 크기 비교** : 누적 막대 차트, 그룹 막대 차트, 선 차트, 영역 차트 등
  - **구성의 비중 비교** : 파이 차트, 100% 누적 막대 차트, 폭포 차트 등
  - **데이터 분포 확인** : 산점도 차트, 히스토그램 등
  - **데이터 관계 비교** : 산점도 차트, 버블 차트 등
 
### [더 정확한 데이터 분석 결과 도출]
* **데이터 분석 결과를 전달**하기 위한 목적으로도 사용
  - 데이터의 정확한 이해를 돕고, 쉽고 빠른 데이터 인사이트 발견에 필수 요소
* 시각화를 활용하여 데이터를 분석하는 **시각적 분석**으로도 활용
  - 데이터 변수, 수치 계산 방식, 차트 유형 등의 조건 변경을 통해 시각적으로 데이터 분석

### 1.2 데이터 시각화 활용 사례

### [시각화 대시보드를 통한 공동의 데이터 활용]
* **시각화 대시보드** : 시각화 차트와 표 등으로 구성된 하나의 보드
  - 주요한 데이터 지표를 모니터링하고 탐색할 때 유용
  - 데이터 조회 기간을 선택하거나 특정 기준으로 데이터를 조회할 수 있는 필터와 같은 상호작용 기능을 제공
  - 복잡한 데이터 분석 기술 없이도 편리하게 데이터를 조회하고 탐색
* 전문 데이터 분석가가 아니어도 쉽게 데이터를 분석 ➜ 기업 및 조직에서 유용하게 활용
  


### [다양한 분양에서의 시각화 활용]
* 데이터 시각화는 데이터가 존재하는 모든 분야에서 활용
* **정부의 시각화 서비스** : 공공데이터 포털 시각화 서비스, 지자체별 공공데이터를 개방하는 다수의 사이트
* **금융권 분야** : 수익 창출을 위한 고객 분석
  - KEB 하나은행의 '하나 빅 인사이트' : 은행 핵심 경영지표, 조직 단위별 영업 실적 모니터링, 고객 특성에 따른 비정형 분석 등 은행 내부 데이터 시각화
* **인공지능 분야** : 복잡한 인공지능 학습 결과에 대한 시각적 분석
  - 클러스터 비전 : 화가 밥 로스 그림 400점의 데이터 셋을 시각적으로 분석
* **언론 분야** : 스토리텔링 기사를 내보낼 때 시각화 차트 이용

### 2.1 시각화 옵션

### [제목]
* **plt.title()** : 그래프 제목을 설정
  - https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html

#### 실습 코드 7-2. 제목 지정

```python
import matplotlib.pyplot as plt

xdata = [2, 3, 4, 5]
ydata = [1, 3, 5, 7]

# 제목 지정을 위해 plt.title() 함수 사용
plt.title('Line Graph')

plt.plot(xdata, ydata)
plt.show()
```

### [범례]
* **plt.legend(labels=['레이블1', '레이블2', ...], loc=위치값)** : 두 개 이상의 데이터를 표현하기 위한 범례 설정
  - plt.plot(label='레이블') 설정한 경우 labels 생략
  - loc : 0 ~ 10까지의 숫자 or  best (0)
  - https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html



#### 실습 코드 7-3. 범례 지정

```python
import matplotlib.pyplot as plt

data1 = [2, 3, 4, 5]
data2 = [8, 6, 4, 2]

plt.title('Legend')

# 1. 범례에 사용될 레이블을 label 속성값으로 지정
plt.plot(data1, label='asc')
plt.plot(data2, label='desc')

# 2. 범례 출력을 위해 plt.legend() 함수 사용
#plt.legend()
plt.legend(labels = ['ASC', 'DESC'], loc=10)  # plt.legend() 함수에 labels 속성을 써서 범례 텍스트 설정도 가능

plt.show()
```

### [x축 및 y축 이름]

### [그래프 선 모양]
* **plt.plot(linestyle='스타일')** : 선 모양 설정
  - 실선 : '-', 파선 : '--', 점쇄선 : '-.', 점선 : ':' 등

### [그래프 선 마커 모양 지정]
* **plt.plot(marker='모양')** : 그래프 선의 점 모양을 설정
  - https://matplotlib.org/stable/api/markers_api.html



#### 실습 코드 7-6. 그래프 선 모양 지정

```python
import matplotlib.pyplot as plt

data1 = [2, 3, 4, 5]
data2 = [8, 6, 4, 2]

# 그래프 선 모양 지정을 위해 plt.plot() 함수의 linestyle 속성값 지정
plt.plot(data1, color='r', marker='h', label='dashed', linestyle='--')
plt.plot(data2, color='g', marker='*', label='dotted', linestyle=':')

plt.legend()
plt.show()
```

### [그림 범위 지정]
* **plt.xlim(), plt.ylim()** : x축, y축 값의 경계 범위(최소값, 최대값) 설정
  - https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlim.html
  - https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylim.html

#### 실습 코드 7-7. 그림 범위 지정

```python
import matplotlib.pyplot as plt
plt.title('X, Y range')
xdata = [5, 10, 15, 20]
ydata = [1, 3, 5, 7]

# 선의 점 모양, 크기를 설정하기 위해 plt.plot() 함수의 marker, markersize 속성값 지정
plt.plot(xdata, ydata, color='b', linestyle='--', marker='o', markersize='10')

# 축의 경계 범위(최소값, 최대값)을 설정하기 위해 plt.xlim(), plt.ylim() 함수 사용
plt.xlim(0,25)
plt.ylim(-2,10)

plt.show()
```

### [데이터 읽어와서 선 그래프 그리기]

#### 실습 코드 7-9. '미세먼지' 엑셀 파일 읽어오기
* **pd.read_excel('파일경로(이름'), [옵션1, 옵션2, ...])** : 엑셀 파일에서 데이터를 읽어와서 데이터프레임 생성
  - 옵션에 들어갈 수 있는 매개변수
    * pd.read_excel( io, **sheet_name=0**, **header=0**, names=None, **index_col=None**, usecols=None, squeeze=False, dtype=None, engine=None, converters=None, true_values=None, false_values=None, **skiprows=None**, nrows=None, na_values=None, keep_default_na=True, na_filter=True, verbose=False, parse_dates=False, date_parser=None, thousands=None, comment=None, skipfooter=0, convert_float=True, mangle_dupe_cols=True)
  - *예) 여러 시트로 나누어진 Excel 파일에서 시트 들고 오려면?*
    * sheet_name 옵션 사용 (sheet_name = "시트명" 또는 sheet_name = index 번호 (시작번호 : 0))

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('fine_dust.xlsx', index_col='area') # 첫번째 열(area)을 행 인덱스 이름으로 지정
data.head()
```

```python
# 2018년 데이터만 읽어와서 data2018 변수에 값을 저장
data2018 = data[2018]
data2018
```

#### 실습 코드 7-10. 기본 그래프 그리기
* **plt.figure(figsize = (x_inch, y_inch))** : 그래프 영역의 크기를 지정
  - https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html
* **plt.grid()** : 그래프 영역의 x, y축에 그리드(격자) 표시
  - **plt.grid(True, axis = 'x')** : x축으로만 격자 표시
  - **plt.grid(True, axis = 'y')** : y축으로만 격자 표시
  - https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html

```python
plt.figure(figsize=(15,4))                      # 그래프 크기 지정 (가로 15인치, 세로 4인치)
plt.plot(data2018, color='b', marker='o')       # 선 그래프 그리기 (data2018에 저장된 미세먼저 데이터, 파란색 선, 원형 마커 사용)
plt.xlabel('area')                              # x축 제목 설정
plt.ylabel('micrometer')                        # y축 제목 설정
plt.title('2018 Fine Dust Line Graph')          # 그래프 제목 설정
plt.grid()                                      # 그래프 영역의 격자 표시 활성화
plt.show()                                      # 그래프를 화면에 출력
```

#### 실습 코드 7-11. 2016~2019년 지역별 미세먼지 선 그래프 (2025년 11월 19일 수업 여기까지)

```python
plt.figure(figsize=(15,4))

# 2016년부터 2029년까지의 지역별 미세먼지 데이터를 모두 선 그래프로 그림
for year in range(2016, 2020):
  chartdata = data[year]
  plt.plot(chartdata, marker='s', label=year)

plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2016~2019 Fine Dust Line Graph')
plt.legend()
plt.grid()
plt.show()
```

### [세로 막대 그래프 그리기]

#### 실습 코드 7-12. 2017년 지역별 미세먼지 세로 막대 그래프

```python
#########################################
# 1) 라이브러리 로드
import pandas as pd
import matplotlib.pyplot as plt

#########################################
# 2) 그래프를 그릴 데이터 준비
data = pd.read_excel('fine_dust.xlsx', index_col='area')
# 2017년 데이터만 읽어와서 data2017 변수(시리즈 타입)에 값을 저장
data2017 = data[2017]

#########################################
# 3) 그래프 종류별 데이터 입력 및 모양 설정
plt.figure(figsize=(15,4))

# 세로 막대 그래프를 그리기 위해 plt.bar() 함수 호출
#  - x 축 : data2017.index (시/도 이름)
#  - y 축 : data2017 (시/도별 미세먼지값)
plt.bar(data2017.index, data2017, color='g')

plt.ylim(35,55)                       # y축 값의 범위 : 35 ~ 55 사이
plt.xlabel('area')                    # x축 제목
plt.ylabel('micrometer')              # y축 제목
plt.title('2017 Fine Dust Bar Graph') # 그래프 제목
plt.grid()                            # 그래프 영역에 격자 표시

#########################################
# 4) 그래프를 화면에 출력
plt.show()
```

### [그룹 세로 막대 그래프 그리기]
* **plt.xticks(ticks, labels, size, rotation, ...)** : x축 눈금 설정하기
   - **ticks** : x축에 표시할 눈금 위치
   - **labels** : 눈금값의 정해주는 이름, 지정한다면 그 이름이 축에 나타난다.
   - **size** : ticks의 크기 조정
   - **rotation** : ticks의 각도 조정 (반시계 방향). 숫자 또는 'horizontal', 'vertical'
   - https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xticks.html

#### 실습 코드 7-13. 2016년~2019년 경기, 인천, 부산, 대구 지역별 미세먼지 그룹 세로 막대 그래프

```python
import numpy as np
index = np.arange(4) # [0, 1, 2, 3]

plt.figure(figsize=(15,4))

# 경기, 인천, 부산, 대구 지역의 2016년~2019년 데이터만 추출하여 data4 변수에 저장
data4 = data.loc['Gyeonggi':'Daegu', 2016:2019]

for year in range(2016, 2020):
    chartdata = data4[year]
    # 그래프의 두께 지정을 위해 width 속성값 설정
    plt.bar(index, chartdata, width=0.2, label=year)
    # 두께 0.2의 막대 그래프 지정하기 위해 막대 그래프가 출력되는 위치를 0.2씩 이동
    index = index + 0.2

plt.ylim(35,55)
plt.xlabel('area')
plt.ylabel('micrometer')

# x축의 눈금을 설정하기 위해 plt.xticks() 함수 사용
plt.xticks(index - 0.5, ['Gyeonggi', 'Incheon', 'Busan', 'Daegu'], rotation=-45)

plt.title('2016~2019 Fine Dust Group Bar Graph')
plt.legend()
plt.show()
```

### [그룹 누적 가로 막대 그래프 그리기]
* 기존 막대 그래프 위에 막대 그래프를 하나 더 그리는 형식
* **plt.barh()** 함수를 사용
  - plt.bar() 함수에서 사용했던 x, y축의 값을 반대로 변경

#### 실습 코드 7-14. 2016년~2019년 지역별 미세먼지 그룹 누적 가로 막대 그래프

```python
plt.figure(figsize=(10,4))
index = np.arange(4)

# 경기, 인천, 부산, 대구 지역의 2016년~2019년 데이터만 추출하여 data4 변수에 저장
data4 = data.loc['Gyeonggi':'Daegu', 2016:2019]
for year in range(2016, 2020):
    chartdata = data4[year]
    # x축에 미세먼지 데이터(chartdata), y축에 시/도 이름 지정하기 위한 index 사용
    # 그래프의 폭 지정을 위해 height 속성값 설정
    plt.barh(index, chartdata, height=0.2, label=year)
    index = index + 0.2

plt.xlim(30,55)
plt.ylabel('area')
plt.xlabel('micrometer')

# y축의 눈금을 설정하기 위해 plt.yticks() 함수 사용
plt.yticks(index - 0.5, ['Gyeonggi', 'Incheon', 'Busan', 'Daegu'])

plt.title('2016~2019 Fine Dust Group Barh Graph')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,4))
index = np.arange(4)

# 경기, 인천, 부산, 대구 지역의 2016년~2019년 데이터만 추출하여 data4 변수에 저장
data4 = data.loc['Gyeonggi':'Daegu', 2016:2019]
for year in range(2016, 2020):
    chartdata = data4[year]
    # x축에 미세먼지 데이터(chartdata), y축에 시/도 이름 지정하기 위한 index 사용
    # 그래프의 폭 지정을 위해 height 속성값 설정
    plt.barh(index, chartdata, label=year)

plt.xlim(30,55)
plt.ylabel('area')
plt.xlabel('micrometer')

# y축의 눈금을 설정하기 위해 plt.yticks() 함수 사용
plt.yticks(index, ['Gyeonggi', 'Incheon', 'Busan', 'Daegu'])

plt.title('2016~2019 Fine Dust Group Barh Graph')
plt.legend()
plt.show()
```

#### 실습 코드 7-15. '2020년 건강검진' 엑셀 파일 읽어오기

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel('health_screenings_2020_1000ea.xlsx')
data.head()
```

#### 실습 코드 7-16. 키(height) 및 몸무게(weight) 산점도 그래프

```python
heightdata = data['height']        # 키 데이터를 들고와서 heightdata 변수에 저장 (시리즈 타입) ==> 산점도 그래프의 X축
weightdata = data['weight']        # 몸무게 데이터를 들고와서 weightdata 변수에 저장 (시리즈 타입) ==> 산점도 그래프의 Y축

# 그래프 영역의 크기 설정
plt.figure(figsize=(10,4))

# 산점도를 그리기 위해 plt.scatter() 함수를 사용
plt.scatter(heightdata, weightdata)

plt.xlabel('height')
plt.ylabel('weight')
plt.title('2020 Health Screenings Scatter Graph')
plt.grid()
plt.show()
```

```python
heightdata = data['systolic']        # 키 데이터를 들고와서 heightdata 변수에 저장 (시리즈 타입) ==> 산점도 그래프의 X축
weightdata = data['diastolic']        # 몸무게 데이터를 들고와서 weightdata 변수에 저장 (시리즈 타입) ==> 산점도 그래프의 Y축

# 그래프 영역의 크기 설정
plt.figure(figsize=(10,4))

# 산점도를 그리기 위해 plt.scatter() 함수를 사용
plt.scatter(heightdata, weightdata)

plt.xlabel('systolic')
plt.ylabel('diastolic')
plt.title('2020 Health Screenings Scatter Graph')
plt.grid()
plt.show()
```

### [그룹 산점도 그래프 그리기]
* 기존 산점도 그래프 위에 산점도 그래프를 추가로 그리는 형식

#### 실습 코드 7-17. 수축기 혈압, 이완기 혈압, 공복 혈당 그룹 산점도 그래프

```python
systolic_data = data['systolic']        # 수축기 혈압 데이터를 들고와서 systolic_data 변수에 저장
diastolic_data = data['diastolic']      # 이완기 혈압 데이터를 들고와서 diastolic_data 변수에 저장
blood_sugar_data = data['blood_sugar']  # 공복 혈당 데이터를 들고와서 blood_sugar_data 변수에 저장

plt.figure(figsize=(10,6))

plt.scatter(systolic_data, diastolic_data, color='r', edgecolor='w', label='systolic*diastolic', s=50, alpha=0.5)
plt.scatter(systolic_data, blood_sugar_data, color='g', edgecolor='w', label='systolic*blood_sugar', s=50, alpha=0.5)
plt.scatter(diastolic_data, blood_sugar_data, color='b', edgecolor='w', label='diastolic*blood_sugar', s=50, alpha=0.5)

plt.xlim(40,180)
plt.ylim(30,300)
plt.title('2020 Health Screenings Group Scatter Graph')
plt.legend()
plt.grid()
plt.show()
```

### [히스토그램 그리기]
* **plt.hist(x, bins='구간개수', color='색상', alpha='투명도', stacked='겹침여부', culmulative='누적여부')** : 히스토그램 그래프 표시
  - https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

#### 실습 코드 7-18. 여성 키 히스토그램

### [그룹 히스토그램 그리기]

#### 실습 코드 7-19. 남성 및 여성 키 그룹 히스토그램

```python
# 데이터프레임 data에서 남성의 성별(gender), 키(height) 데이터만 가져와서 male_data에 저장
male_data = data.loc[data.gender==1,['gender','height']]
# 데이터프레임 data에서 여성의 성별(gender), 키(height) 데이터만 가져와서 female_data에 저장
female_data = data.loc[data.gender==2,['gender','height']]

plt.figure(figsize=(10,6))

# plt.hist() 함수를 이용하여 히스토그램 그래프 표시
#  - 구간의 개수는 7개로 지정
#  - 두 히스토그램의 데이터 구간이 겹치므로 겹치는 그래프를 표시하기 위해 투명도(alpha) 설정
plt.hist(male_data['height'], bins=7, alpha=0.5, label='Male', color='g')
plt.hist(female_data['height'], bins=7, alpha=0.5, label='Female', color='r')

plt.xlim(130,195)
plt.xlabel('height')
plt.ylabel('frequency')
plt.title('2020 Health Screenings Male & Female Height Group Histogram')
plt.legend()
plt.grid()
plt.show()
```

### [그룹 누적 히스토그램 그리기]
* **plt.hist(stacked = True)** 설정 : 다수의 데이터를 겹쳐서 표현
* **plt.hist(cumulative = True)** 설정 : 앞 구간의 데이터를 누적에서 그래프 표현
#### 실습 코드 7-20. 남성 및 여성 키 그룹 누적 히스토그램

```python
import numpy as np

# 남성의 키 데이터와 여성의 키 데이터를 가져와서 ndarray를 생성한 후 height 변수에 저장
height = np.array([male_data['height'], female_data['height']], dtype=object)

plt.figure(figsize=(10,6))

# stacked = True로 설정하여 남성 키와 여성 키 히스토그램 그래프를 겹쳐서 표현
plt.hist(height, bins=7, label=['Male','Female'], alpha = 0.5, stacked=True)

plt.xlim(130,195)
plt.xlabel('height')
plt.ylabel('frequency')
plt.title('2020 Health Screenings Male & Female Height Stacked Histogram')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))

# cumulative = True로 설정하여 앞 구간의 데이터를 계속 누적하여 그래프 표현
plt.hist(height, bins=7, label=['Male','Female'], alpha = 0.5, cumulative=True)

plt.xlim(130,195)
plt.xlabel('height')
plt.ylabel('frequency')
plt.title('2020 Health Screenings Male & Female Height Stacked Histogram')
plt.legend()
plt.show()
```

### [상자수염 그래프 그리기]

#### 실습 코드 7-21. 남성 및 여성 키 상자수염 그래프

```python
import numpy as np

# 남성의 키 데이터와 여성의 키 데이터를 가져와서 ndarray를 생성한 후 height 변수에 저장
height = np.array([male_data['height'], female_data['height']], dtype=object)

plt.figure(figsize=(8,5))

# plt.boxplot() 함수로 상자수염 그래프 표시
#  - 남자(Male), 여자(Female) 두 그룹별 키 데이터에 대한 상자수염 그래프 표시
plt.boxplot(height, labels=['Male','Female'])

plt.ylim(130,200)
plt.xlabel('gender')
plt.ylabel('height')
plt.title('2020 Health Screenings Male & Female Height Box Plot')
plt.show()
```

#### 실습 코드 7-22. 남성 및 여성 키 가로 상자수염 그래프

```python
plt.figure(figsize=(8,5))

# plt.boxplot() 함수로 상자수염 그래프 표시
#  - vert = False : 가로 상자수염 그래프로 설정 (True : 세로 상자수염 그래프, 기본값)
plt.boxplot(height, labels=['Male','Female'], vert=False)

plt.xlim(130,200)
plt.ylabel('gender')
plt.xlabel('height')
plt.title('2020 Health Screenings Male & Female Height Box Plot')
plt.show()
```
