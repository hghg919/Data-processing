## 6. 선택문 (조건문)

* **순차문(Sequaltial Statements)**
  - 먼저 나타나는 코드가 먼저 실행되는 구조 (위에서 아래로 순차적으로 실행)
* **제어문(Control Statements)**
  - 조건을 만족하는 동안 특정 구간을 한 번 또는 여러 번 반복하여 실행하도록 제어
  - *선택문(Conditional Statements), 반복문(Loop Statements)*
 
### 6.1 파이썬 선택문 개요

* 실행을 달리하는 여러 개의 실행문이 있음
* 특정한 조건에 따라서 실행됨
* 조건식은 True 혹은 False 반환
* 선택문의 형태 : if, if ~ else, if ~ elif, if ~ elif ~ else, if ~ in ~ else 등

### [파이썬 선택문의 규칙]
* if문의 조건식들은 논리 연산이 가능한 문장으로 참 거짓 판별이 가능하도록 작성해야 한다.
* if문의 조건식 끝에는 항상 :(콜론)이 있어야 한다.
* 실행 코드는 반드시 공백([Spacebar] 또는 [Tab])으로 들여쓰기(ident)를 하여 if 문에 포함되는(종속) 코드로 작성해야 한다.

## 7. 반복문과 기타 제어문

* 특정 문장을 여러 번 반복하여 실행하는 구조
* 범위의 횟수나 조건식이 만족하는 동안 여러 번 반복하여 실행
* 반복문의 종류 : for 문, while 문
  - **for 문** : 범위의 횟수 동안 특정 문장을 여러 번 반복하여 실행
  - **while 문** : 조건식의 결과가 True면 특정 문장을 여러 번 반복 실행하다가 조건식이 False가 되면 반복 종료
 
### 7.1 파이썬 반복문 개요

### [파이썬 반복문의 규칙]
* for 문과 while 문의 조건식 끝에는 항상 :(콜론)이 있어야 한다.
* 실행 코드는 반드시 공백([Spacebar] 또는 [Tab])으로 들여쓰기(indent)를 하여 for 문과 while 문에 포함되는(종속) 코드로 작성해야 한다.

### [for 문의 규칙]
* 제어변수와 특정 범위를 활용하여 반복 실행한다.
* 특정 범위는 range() 함수, 리스트, 문자열 등을 이용하여 지정한다.

### [while 문의 규칙]
* 선택문처럼 조건식의 결과가 참 또는 거짓인지에 따라 특정 문장을 반복 실행하거나 실행하지 않는 조건적 반복 실행 구조이다.
* while 문의 조건식들은 논리 연산이 가능한 문장, 즉 참 또는 거짓 판별이 가능한 것이어야 한다.

### 7.4 기타 제어문

### [반복문을 제어하는 키워드 : break, continue, pass]
* **break** : 반복 구간을 더 이상 실행하지 않고 반복문을 강제 종료
* **continue** : 반복 구간의 남은 명령문을 실행하지 않고 반복문 처음으로 이동하여 계속 실행 **(반복문 실행 종료 X)**
* **pass** : 아무 것도 실행하지 않는 문장. 반복문이나 임시 함수 작성 시 기본 틀만 표현하기 위해 임시로 작성하는 코드

#### 실습 코드 3-46. 무한 반복 while 문 안에서 break를 이용한 강제 종료

### 7.5 중첩 반복문

* 반복문 내부에 또 다른 반복문이 표현된 형태
* 한 번의 반복문으로 수행하기 어려운 복잡한 문제 해결에 사용

#### 실습 코드 3-49. 0을 입력하기 전까지 단을 입력받아 구구단 출력

````python
while True:
    num = int(input('단 입력(종료:0) '))
    if num == 0:
        break
    a = 1
    while a <= 9:
        print(f'{num} × {a} = {num*a}')
        a += 1
print('구구단 프로그램 종료')
````

## 8. 함수

### 8.1 함수란?

* 특정 작업을 위해 서로 관련되어 있는 일련의 명령문


### [함수를 사용하는 이유]
* 코드의 양을 줄일 수 있고, 이로 인해서 가독성이 좋아진다.
* 간결해진 코드는 전체의 기능을 이해하기에 수월하다.
* 프로그램의 흐름을 파악하기 쉽기에 관리하기 쉽다.
* 필요로 하는 프로그램이나 다른 프로그래머에게 재사용이 가능하여 매우 효율적으로 사용할 수 있다.

### [함수의 종류]
* **내장 함수(Built-in function)** : 프로그램에 이미 지정된 함수
* **사용자 정의 함수(User-defined function)** : 사용자가 필요에 의해 직접 정의하는 함수

### 8.2 사용자 정의 함수

* 함수 정의의 기본 구조


* 함수 이름의 규칙
  - 알파벳, 숫자, 밑줄 문자(_)만 사용 가능
  - 함수의 첫 글자는 반드시 알파벳 또는 밑줄 문자만 사용 가능
  - 공백 사용 불가
  - 예약어(print, def 등) 사용 불가


#### 실습 코드 3-51. 두 개의 수를 입력받아 큰 수 결정하는 함수의 예

````python
def max1(num1,num2):
    if num1>num2:
        return num1
    else:
      return num2
a=int(input('숫자1: '))
b=int(input('숫자2: '))
print(f'큰수: {max1(a,b)}')
````

#### 실습 코드 3-52. 함수를 활용한 3, 6, 9 게임 예

````python
def game(num):
    a=num//10
    b=num%10    
    if a==3 or a==6 or a==9:
        return '통과'
    elif b==3 or b==6 or b==9:
        return '통과'
    else:
        return '다음 수'

num=int(input('숫자: '))
print(game(num))
````

### 8.3 람다(lambda) 함수

* 이름 없는 익명 함수 (ananymous function : 일회용의 간단한 함수)
* 식 형태로 되어 있어 람다 표현식(lambda expression)이라고도 부름

### [람다 표현식 작성]


````python
# 숫자를 전달 받은 뒤 5를 더해서 반환하는 함수 (plus_five)
def plus_five(x):
  return x + 5

print('10과 5의 합 :', plus_five(10))
````

````python
# plus_five() 함수의 람다 표현식
lambda x: x + 5
````

````python
# 람다 표현식의 익명 함수를 호출하기 위해 plus_five 변수에 할당
plus_five = lambda x: x + 5
print('10과 5의 합 :', plus_five(10))
````

````python
# 인라인(inline) 람다 함수와 인자를 이용
print('10과 5의 합 : ', (lambda x: x + 5)(10))
````



* 람다 표현식 안에는 새 변수를 생성할 수 없음
* 함수 내부에 변수가 필요하면 사용자 정의 함수를 작성해야 함

### [람다 표현식을 인수로 사용]
* 함수의 인수 부분에서 간단하게 함수를 만들기 위해 사용
* **map() 함수**와 연동하여 사용
  - 여러 개의 데이터를 받아서 각각의 요소에 함수를 적용한 결과를 반환하는 내장 함수
  - 리스트, 튜플 등의 반복 가능한 객체를 입력으로 받을 수 있음
  - 기본 문법 : map(적용시킬 함수, 반복가능 객체)

````python
def plus_five(x):
  return x + 5

list(map(plus_five, [30, 20, 10]))
````

````python
list(map(lambda x: x + 5, [30, 20, 10]))
````

## 9. 파일 입출력

### 9.1 파일입출력 기본

### [파일 개념]
* 컴퓨터의 저장 장치 내에 데이터를 저장하기 위해 사용하는 논리적인 단위

### [파일 입출력 기본 과정]
1. 파일 열기

#### 실습 코드 3-53. 파일 읽기 : read() 함수

#### 실습 코드 3-55. 파일 읽기 : readline() 함수 (2)
* 여러 줄을 포함하는 파일에서 내용을 끝까지 읽어오는 코드

#### 실습 코드 3-57. 파일 읽기 : readlines() 함수 (2)

````python
f=open('name.txt', encoding='utf-8')
data=f.readlines()
for line in data:
    print(line,end='')
f.close()
````

### [with ~ as 구문]
* 문법을 명확하게 만들고 예외를 손쉽게 처리하는 방법
* with 구문 이용 시 try-except-finally를 대신하여 더욱 쉽게 사용 가능
* **파일 쓰기가 완료되면 자동으로 f.close()가 수행되기 때문에 finally 절을 사용할 필요가 없음**

````python
f = open('name.txt', 'r', encoding='utf-8')
try :
    data = f.readlines()
    print(data, end = '')
finally :
    f.close()
````

````python
with open('name.txt', encoding='utf-8') as f:
    data = f.readlines()
    print(data, end = '')
````

### [텍스트 파일 쓰기]
* write(), writelines() 함수를 이용하여 파일에 내용 저장

#### 실습 코드 3-58. write() 함수
* 동아리 회원의 이름을 사용자로부터 입력받아 파일에 저장

````python
f=open('club.txt','w')

while True:
    name=input('이름: ')
    if not name:
        break
    f.write(name)
f.close()
````

* write() 함수는 자동 개행(다음줄 넘김)이 포함되지 않기 않음
* 파일에 줄 단위로 저장하고 싶으면 개행 문자(\n)를 포함해야 함

````python
f=open('club.txt','w')

while True:
    name=input('이름: ')
    if not name:
        break
    f.write(name+'\n')
f.close()
````

#### 실습 코드 3-59. 파일 추가 모드 : write() 함수
* 기존 파일 내용을 유지하면서 새로운 값을 추가

### 9.4 파일 오류 처리
* 파일 입출력 시 다양한 예외 상황 발생 가능
  - 존재하지 않는 파일을 열려고 시도
  - 읽기 모드로 열었는데 파일에 내용을 쓰려고 시도 등
* open() 함수로 파일을 열려고 할 때 파일이 없으면 오류 발생
  - **os.path.exists(파일명), os.path.isfile(파일명) 함수 사용**
  - **예외 처리 구분(try ~ except) 사용**

#### 실습 코드 3-60. 현재 디렉토리에 읽고자 하는 파일이 없는 경우 오류 메시지

````python
f=open('test.txt','r')

while True:
    name=input('이름: ')
    if not name:
        break
    f.write(name+'\n')
f.close()
````

#### 실습 코드 3-61. 파일 오류 처리 : os.path.exists() 사용
* 사용자로부터 읽으려는 파일명을 입력받아 파일이 존재할 경우 파일의 모든 내용을 읽어서 화면에 출력
* 파일이 존재하지 않으면 오류 메시지 출력

````python
import os.path

fname=input('파일명: ')
fname=fname+'.txt'
if os.path.exists(fname): # os.path.isfile(fname)
    f=open(fname,'r')
    data=f.read()
    print(data,end='')
    f.close()
else:
    print('{} 파일이 없습니다.'.format(fname))
````

#### 실습 코드 3-62. 파일 오류 처리 : try ~ except 사용
* 파일이 존재할 경우 파일의 모든 내용을 읽어서 화면에 출력
* 파일이 존재하지 않으면 except 구문 내의 오류 메시지 출력

````python
fname=input('파일명: ')
fname=fname+'.txt'
try:
    f=open(fname)
    data=f.read()
    print(data,end='')
    f.close()
except:
    print('{} 파일이 없습니다.'.format(fname))
````

### [파이썬 예외처리 코드 기본 구조]

```python
try :
    # 예외가 일어날 것 같은 코드, 즉 위험한 코드
except :
    # 예외가 있을 때 실행될 코드
else :
    # 예외가 없을 때 실행될 코드
finally :
    # 예외와 상관없이 무조건 실행될 코드
```
