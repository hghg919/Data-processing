### 1.1 변수의 개념

* 변수(Variable)는 쉽게 변하는 수로, 프로그램이 동작하면서 어떤 상황 혹은 상태에 따라 변화하는 어떤 자료(데이터)를 담아 두기 위해 사용하는 개념이다.

### [% 형식지정자를 이용한 출력]

#### 실습코드 3-16. %d 형식지정자를 이용하여 출력

```python
a = int(input('숫자1 입력: '))
b = int(input('숫자2 입력: '))
print('%d + %d = %d' % (a, b, a+b))
```

## 4. 컬렉션(Collection) 자료형

### 4.1 컬렉션 자료형 개념과 필요성

* 여러 개의 값을 하나의 변수에 담을 수 있다.
* 변수 안에 공간을 여러 개 가진다.
* 변수 안에 서로 다른 공간을 찾는 방법이다.



### 4.2 리스트(list) 자료형

* 대괄호 [] 안에 서로 다른 자료형의 값을 콤마(,)로 구분해 하나 이상 저장할 수 있는 컬렉션 자료형
* 파이썬 프로그래밍 언어 내에서 가장 많이 쓰이는 구조
* 많은 양의 데이터들을 한 번에 모아 효율적으로 처리하고 저장

#### 실습 코드 3-21. 일반 변수 사용

```python
hap=0
a=int(input('숫자1: '))
b=int(input('숫자2: '))
c=int(input('숫자3: '))
d=int(input('숫자4: '))

hap=a+b+c+d

print(f'합계: {hap}')
```

#### 실습 코드 3-22. 리스트 변수 사용

```python
a=[10,10,10,10]
a[0]=int(input('숫자1: '))
a[1]=int(input('숫자2: '))
a[2]=int(input('숫자3: '))
a[3]=int(input('숫자4: '))

hap=a[0]+a[1]+a[2]+a[3]

print(f'합계: {hap}')
```

## 5. 리스트 내포(List Comprehension)

* 원하는 자료들을 조회 또는 추출하여 리스트로 변환하는 표현식
* 조건식을 이용하여 같은 연산을 전체 항목이나 일부 항목에 적용

### 5.1 리스트 내포 처리 과정

1. 항목들에서 순차적으로 하나씩 꺼내 온다. 리스트에서 꺼내 올 항목이 없으면 종료한다.
2. 조건식을 적용하여 해당 조건에 맞는 항목은 추출하고, 조건에 맞지 않으면 무시한다.
3. 2에서 추출된 항목은 리스트에 추가한다.
4. 1로 이동하여 반복한다.

### [기본적인 리스트 내포 형식]


* 리스트 내포 적용 예제

```python
list1=[-40,-20,0,20,40]
[x for x in list1]
```

```python
list1=[-40,-20,0,20,40]
[x+5 for x in list1]
```

```python
list1=[-40,-20,0,20,40]
[x/5 for x in list1]
```

```python
list1=[-40,-20,0,20,40]
[x**2 for x in list1]
```

### [조건 있는 리스트 내포 형식]


* 조건을 지정하여 표현식의 오류 방지 (예: 분모 0이 되는 항목 배제)

```python
# list1의 요소 중 0이 분모로 입력되어 "division by zero" 에러 발생
list1=[-40,-20,0,20,40]
[1/x for x in list1]
```

```python
list1=[-40,-20,0,20,40]
[1/x for x in list1 if x>0]
```

* 컬렉션 자료형, range() 함수를 이용하여 항목 생성 가능

```python
[num * 5 for num in range(3)]
```

```python
[num * 5 for num in range(10) if num % 2 == 0]
```

* 알파벳 소문자 리스트 생성

```python
alpha = [chr(n) for n in range(97,97+26)]
print(alpha)
```
