**5. 다음과 같이 일주일 동안의 온도 데이터를 나타내는 배열 temperatures가 있다. 최대 온도와 최소 온도를 계산하고, 온도 변화의 표준편차를 구하는 코드를 작성하시오. (10점)**

<<출력 결과>>

최대 온도: 25.0  
최소 온도: 19.5  
온도 변화의 표준편차: 1.9430672155336348

```python
# 셀에 작성된 코드를 먼저 실행 후 코드 작성할 것
temperatures = np.array([22.5, 24.0, 21.0, 19.5, 23.5, 20.0, 25.0])
```

```python
# 여기에 코드 작성
print("최대 온도 : ", end='')
print(temperatures.max())
print("최소 온도 : ", end='')
print(temperatures.min())
print("온도 변화의 표준편차 : ", end='')
print(temperatures.std())
```

시리즈(Series) : Index 레이블을 갖는 동일한 유형의 데이터를 저장하는 1차원 배열, 엑셀의 하나의 열과 같다.

딕셔너리의 키 값들이 열의 이름이 되며 밸류들이 리스트가 된다.

**8. 생성된 데이터프레임 변수를 다른 변수에 복사해서 수정 작업을 할 때 주의해야 한다. 어떤 문제가 발생할 수 있는지 얕은 복사, 깊은 복사를 이용하여 설명하시오. (10점)**

df3 = df -> 얕은 복사 (Shallow Copy) : df와 df3는 메모리를 공유한다. df3의 데이터를 수정시 원본인 df의 데이터또한 수정된다.
df3 = df.copy -> 깊은 복사 (Deep Copy) : df와 df3는 메모리 공유하지 않는다. 따라서 df3의 데이터를 수정해도 df는 영향을 받지 않는다.
